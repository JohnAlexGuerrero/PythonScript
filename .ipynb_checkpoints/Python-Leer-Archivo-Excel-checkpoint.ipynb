programa para leer un archivo excel en python

```python
import pandas as pd
```

```python
# Lectura del archivo Excel
df = pd.read_excel('inventario_final.xlsx')
```

```python
df
```

```python
#conteo de numero de filas en el dataframe
len(df)
```

```python
for i in range(len(df)):
    new_product = {
        "id": df.iloc[i]['id'],
       "name": df.iloc[i]['name'],
        "codebar": df.iloc[i]['codebar'],
        "qty": df.iloc[i]['qty']
    }
    
    print(new_product)
```
